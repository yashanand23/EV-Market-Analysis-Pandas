```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure all data displays without truncation
pd.set_option('display.max_columns', None)
```

### Data Loading

```python
import pandas as pd

def load_data():
    # Load datasets from CSV files
    ev_main = pd.read_csv(r'C:\Users\Coditas-Admin\Documents\GitHub\Pandas-Assignment\data\ev_main.csv')
    ev_efficiency = pd.read_csv(r'C:\Users\Coditas-Admin\Documents\GitHub\Pandas-Assignment\data\ev_efficiency.csv')
    ev_manufacturing = pd.read_csv(r'C:\Users\Coditas-Admin\Documents\GitHub\Pandas-Assignment\data\ev_manufacturing.csv')
    ev_market_share = pd.read_csv(r'C:\Users\Coditas-Admin\Documents\GitHub\Pandas-Assignment\data\ev_market_share.csv')
    
    # Check if 'launch_date' exists and convert it if it does
    if 'launch_date' in ev_main.columns:
        ev_main['launch_date'] = pd.to_datetime(ev_main['launch_date'], errors='coerce')
    else:
        print("'launch_date' column not found!")
    
    return ev_main, ev_efficiency, ev_manufacturing, ev_market_share

# Load all data into variables
ev_main, ev_efficiency, ev_manufacturing, ev_market_share = load_data()

# Quick overview of loaded data
print("EV Main Data:\n", ev_main.head())
print("\nEV Efficiency Data:\n", ev_efficiency.head())
print("\nEV Manufacturing Data:\n", ev_manufacturing.head())
print("\nEV Market Share Data:\n", ev_market_share.head())
```

## Series Operations

```python
# Create a Pandas Series containing EV brands
brands = pd.Series(['Tesla', 'Nissan', 'Chevrolet', 'BMW', 'Ford'])

# Perform basic operations like indexing and string methods
print("Brand at index 1:", brands[1])
print("Uppercase brands:\n", brands.str.upper())

# Calculate and display the mean length of the brand names
mean_length = brands.str.len().mean()
print("Mean length of brand names:", mean_length)
```

### Working with DataFrames

```python
# Cell 3: Task 2 - Working with DataFrames

# Display DataFrame information
print("DataFrame Info:")
print(ev_main.info())

# Display DataFrame's axes
print("DataFrame Axes:", ev_main.axes)

# Access specific values (Example: first 5 rows, first 3 columns)
print("First 5 rows, first 3 columns:\n", ev_main.iloc[:5, :3])

# Apply a filter to show EVs with a range over 250 miles
high_range_ev = ev_main[ev_main['Range'] > 250]
print("EVs with range over 250 miles:\n", high_range_ev)

# Calculate the average price of EVs
avg_price = ev_main['Price'].mean()
print("Average price of EVs:", avg_price)
```

### Handling Duplicates and NA Values

```python
# Cell 4: Task 3 - Handling Duplicates and NA Values

# Remove duplicates based on the 'Model' column
ev_main.drop_duplicates(subset='Model', inplace=True)

# Identify NA values in the DataFrame
na_counts = ev_main.isna().sum()
print("NA values in each column:\n", na_counts)

# Fill NA values in the 'Sales' column with the median sales value
ev_main['Sales'] = ev_main['Sales'].fillna(ev_main['Sales'].median())

# Forward-fill for remaining NA values
ev_main = ev_main.ffill()  # Forward fill all remaining NA values

print("Remaining NA values after handling:\n", ev_main.isna().sum())
```

### Advance DataFrame Operations

```python
# Cell 5: Task 4 - Advanced DataFrame Operations

# Merge ev_main with ev_efficiency on 'Model'
merged_data = pd.merge(ev_main, ev_efficiency, on='Model', how='inner')

# Left join with ev_manufacturing on 'Brand'
merged_data = pd.merge(merged_data, ev_manufacturing, on='Brand', how='left')

# Split and concatenate DataFrames based on Price > 50000
high_price_data = merged_data[merged_data['Price'] > 50000]
low_price_data = merged_data[merged_data['Price'] <= 50000]
concatenated_data = pd.concat([high_price_data, low_price_data])
print("Concatenated Data Matches Original:", concatenated_data.shape == merged_data.shape)

# Apply custom functions for Efficiency_Score and Value_Score
# Use 'Efficiency' (capital E) instead of 'efficiency'
merged_data['Efficiency_Score'] = merged_data['Efficiency'].apply(lambda x: 'High' if x > 3 else 'Low')
merged_data['Value_Score'] = (merged_data['Range'] * merged_data['Efficiency']) / merged_data['Price']
print("Data with Efficiency and Value Scores:\n", merged_data[['Brand', 'Model', 'Efficiency_Score', 'Value_Score']])

# Create a pivot table and melt
pivot_table = merged_data.pivot_table(values='Price', index='Brand', columns='Efficiency_Score', aggfunc='mean')
print("Pivot Table of Average Price by Brand and Efficiency Score:\n", pivot_table)

melted_data = pd.melt(merged_data, id_vars=['Brand', 'Model'], value_vars=['Range', 'Price', 'Sales'])
print("Melted Data (Long Format):\n", melted_data.head())
```

### Working with Vectors

```python
# Cell 6: Task 5 - Working with Vectors
import numpy as np

# Extract 'Range' and 'Price' columns as NumPy arrays
range_array = ev_main['Range'].to_numpy()
price_array = ev_main['Price'].to_numpy()

# Calculate correlation coefficient between range and price
correlation = np.corrcoef(range_array, price_array)[0, 1]
print("Correlation between range and price:", correlation)

# Calculate price per mile of range
price_per_mile = price_array / range_array
print("Price per mile of range:\n", price_per_mile)
```

### Exploratory Data Analysis with Pandas

```python
# Generate descriptive statistics
print("Descriptive statistics:\n", ev_main.describe())

# Scatter plot of Range vs Price
plt.figure(figsize=(8, 6))
plt.scatter(ev_main['Range'], ev_main['Price'], alpha=0.5)
plt.xlabel('Range')
plt.ylabel('Price')
plt.title('Range vs Price Scatter Plot')
plt.show()

# Bar plot of average Efficiency by Brand
# Corrected 'efficiency' to 'Efficiency'
avg_efficiency_by_brand = merged_data.groupby('Brand')['Efficiency'].mean()
avg_efficiency_by_brand.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.ylabel('Average Efficiency')
plt.title('Average Efficiency by Brand')
plt.show()

# Box plot of Price distribution by Brand
plt.figure(figsize=(10, 6))
merged_data.boxplot(column='Price', by='Brand', grid=False, rot=45)
plt.title('Price Distribution by Brand')
plt.suptitle('')
plt.show()

# Pie chart of market share based on Sales
market_share = ev_main.groupby('Brand')['Sales'].sum()
market_share.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(8, 8))
plt.ylabel('')
plt.title('Market Share based on Sales')
plt.show()
```

```python
### Advanced Analysis

# Check columns to verify the structure of merged_data
print("Columns in merged_data:", merged_data.columns)

# Group data by 'Brand' and another available column if 'primary_manufacturer' does not exist.
# Replace 'primary_manufacturer' with an existing column (if applicable).
grouped_data = merged_data.groupby(['Brand']).agg(
    mean_range=('Range', 'mean'),
    mean_price=('Price', 'mean'),
    mean_efficiency=('Efficiency', 'mean')
)
print("Grouped Data:\n", grouped_data)

# Bin 'Price' into categories
price_bins = pd.cut(merged_data['Price'], bins=[0, 30000, 50000, 100000], labels=['Budget', 'Mid-range', 'Luxury'])
merged_data['Price_Category'] = price_bins
price_distribution = merged_data['Price_Category'].value_counts()
print("Price Category Distribution:\n", price_distribution)

# Identify top 5 models based on 'Value_Score'
top_5_models = merged_data.nlargest(5, 'Value_Score')[['Brand', 'Model', 'Value_Score']]
print("Top 5 models by Value Score:\n", top_5_models)

# Calculate percentage of High vs Low efficiency vehicles for each Brand
# Ensure 'Efficiency_Score' column exists
efficiency_distribution = merged_data.groupby('Brand')['Efficiency_Score'].value_counts(normalize=True) * 100
print("Percentage of High vs Low efficiency vehicles for each Brand:\n", efficiency_distribution)

# Perform time-based analysis
merged_data['Launch_Year'] = merged_data['Launch_Date'].dt.year
yearly_sales = merged_data.groupby('Launch_Year')['Sales'].sum()
print("Yearly Sales Trend:\n", yearly_sales)
```
